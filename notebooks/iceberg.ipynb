```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("openlineage-demo")
    .getOrCreate()
)
```

```python
!wget https://github.com/plotly/datasets/raw/refs/heads/master/2015_flights.parquet
```

```python
spark.sql("show databases").show()
```

```python
flights = spark.read.parquet("2015_flights.parquet")
```

```python
flights \
  .filter("distance < 150") \
  .write \
  .mode("overwrite") \
  .saveAsTable("short_flights")
```

```python
flights.write.mode("overwrite").format("iceberg").saveAsTable("iceberg_flights")
```

```python
spark \
  .read \
  .table("iceberg_flights") \
  .filter("distance < 150") \
  .write \
  .format("iceberg") \
  .mode("overwrite") \
  .saveAsTable("short_flights")
```

```python
flights.write.mode("overwrite").format("parquet").saveAsTable("parquet_flights")
```
